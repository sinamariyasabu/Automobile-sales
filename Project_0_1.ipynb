##Histogram

Histogram refers to the frequency of occurrence of variables in an interval

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
df=pd.read_csv('/content/Auto Sales data.csv')
plt.figure(figsize=(6,3.2),dpi=150)
plt.rc('xtick', labelsize=6)
plt.rc('ytick', labelsize=8)
sns.distplot(df['DAYS_SINCE_LASTORDER'])
plt.show()
```

##Pie plot

A pie plot is a proportional representation of the numerical data in a column.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
df = pd.read_csv("/content/Auto Sales data.csv")
df.groupby(['STATUS']).sum().plot( kind='pie', y='QUANTITYORDERED',autopct='%1.0f%%')
plt.show()
```

##Heat Maps

Heat Maps is a type of plot which is necessary when we need to find the dependent variables. One of the best way to find the relationship between the features can be done using heat maps.

```python
import seaborn as sns
sns.heatmap(df.corr(numeric_only="SALES"),cmap='Greens',annot=True)
```
